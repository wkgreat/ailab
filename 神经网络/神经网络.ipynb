# 数据集读取函数

```python
import numpy as np

def load_mnist_images(filename):

    with open(filename, 'rb') as f:

        magic = int.from_bytes(f.read(4), 'big')
        num = int.from_bytes(f.read(4), 'big')
        rows = int.from_bytes(f.read(4), 'big')
        cols = int.from_bytes(f.read(4), 'big')

        print("magic:", magic)
        print("num:", num)
        print("size:", rows, cols)

        data = np.frombuffer(f.read(), dtype=np.uint8)

        images = data.reshape(num, rows, cols)

        return images
    
def load_mnist_labels(filename):

    with open(filename, 'rb') as f:

        magic = int.from_bytes(f.read(4), 'big')
        num = int.from_bytes(f.read(4), 'big')

        print("magic:", magic)
        print("num:", num)

        labels = np.frombuffer(f.read(), dtype=np.uint8)

        return labels
```

```python
train_images = load_mnist_images("./data/minist/train-images-idx3-ubyte")
train_labels = load_mnist_labels("./data/minist/train-labels-idx1-ubyte")
test_images = load_mnist_images("./data/minist/t10k-images-idx3-ubyte")
test_labels = load_mnist_labels("./data/minist/t10k-labels-idx1-ubyte")

print(len(train_images))
```
